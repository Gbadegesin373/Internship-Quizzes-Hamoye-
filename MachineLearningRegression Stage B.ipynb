Question 13: Normalize by minmax scaler
remove [date and lights]
target variable is appliances
Find the Mean absolute error in 2 dp

```python
energy_df.head()
```

```python
new_energy_df = energy_df.drop("lights", axis = 1)
```

```python
num_attribs = new_energy_df.columns
```

```python
num_attribs
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
```

```python
#Creating a pipeline
num_pipeline = Pipeline([
    ("MinMax", MinMaxScaler())
])
```

```python
from sklearn.compose import ColumnTransformer

col_trans = ColumnTransformer(transformers=[
    ("num_pipeline", num_pipeline, num_attribs),
],
                             remainder = "drop",
                             n_jobs = -1)
```

```python
from sklearn.pipeline import FeatureUnion
```

```python
full_pipeline = FeatureUnion(transformer_list = [
    ("col_trans", col_trans)
])
```

```python
#Let us run the whole pipeline simply
transformed_energy = full_pipeline.fit_transform(new_energy_df)
```

```python
transformed_energy
```

```python
#converting the numpy array to DataFrame
transformed_energy_df = pd.DataFrame(transformed_energy)
```

```python
transformed_energy_df
```

```python
features_df = transformed_energy_df.drop(columns = [0])
```

```python
target_df = transformed_energy_df[0]
```

```python
target_df.shape
```

```python
#spliting into training and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, target_df, test_size = 0.3, random_state = 42)
```

```python
#fitting a linear model
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
```

```python
#Testing of model using test set
y_pred = lin_reg.predict(x_test)
```

```python
#evaluating the MSE
#Let measure this regression model's RMSE
from sklearn.metrics import mean_squared_error
lin_mse = mean_squared_error(y_test, y_pred)
lin_rmse = np.sqrt(lin_mse)
```

```python
np.round(lin_mse,2)
```

Question 17 - Obtain the feature weights from your linear model above


```python
lin_reg.intercept_
lin_reg.coef_
```

```python
def feature_weights(model, features):
    model_coef = model.coef_
    weights = pd.Series(model_coef, features).sort_values(ascending = False)
    weights_df = pd.DataFrame(weights, columns = ["Weight"])
    
    return weights_df
```

```python
features = features_df.columns
feature_weights(lin_reg, features)
```

Answer is RH_2 (lowest) and RH_1 (highest)
