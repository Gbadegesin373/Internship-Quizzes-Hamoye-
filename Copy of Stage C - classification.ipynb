# Some preprocessing

```python
#let us convert the stabf into numericals
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
#Let us separate the data into the predictors and target variables
X = df.drop("stabf_unstable", axis = 1)
y = df["stabf_unstable"]
```

```python
#Let us standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#splitting the data into test and train set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = .2, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

# Modelling
Random Forest Classifier - Question 14

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier(random_state = 1)
rf_clf.fit(X_train, y_train)
```

```python
#predictions
rf_clf_predictions = rf_clf.predict(X_test)
```

```python
#evaluating the predictions of the X_test using some metrics
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score

rf_clf_accuracy = accuracy_score(y_test, rf_clf_predictions)
print("\nClassification Report:_______________RF_CLF")
print(classification_report(y_test, rf_clf_predictions))

print("Accuracy: {:.4f}".format(rf_clf_accuracy))
```

```python
param_randomized = [
    {'n_estimators': [50, 100, 300, 500, 1000], "min_samples_split": [2,3,5,7,9],
     "min_samples_leaf": [1, 2, 4, 6, 8], "max_features": ["sqrt", "log2", " auto", None]
}]
random_search = RandomizedSearchCV(estimator = ext_clf, param_distributions = param_randomized, cv =5
                                   , n_iter = 10, scoring = "accuracy", n_jobs = -1,
                                   verbose = 1, random_state = 1)
random_search.fit(X_train, y_train)
```
